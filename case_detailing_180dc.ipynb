```python
section_vectors = model.encode(law_sections)
dimension = section_vectors.shape[1]  # Dimensionality of sentence embeddings
index = faiss.IndexFlatL2(dimension)   # Index with L2 (Euclidean distance)
index.add(section_vectors)             # Add vectors to the index

# 3. Perform a search (query)
query = "Attempt to commit a crime"
query_vector = model.encode([query])   # Convert query to vector

# Search the index for the closest vector (top 3 matches)
k = 3  # Number of results to retrieve
distances, indices = index.search(query_vector, k)

# 4. Show the results
print(f"Query: {query}\n")
for idx in indices[0]:
    print(f"Section: {law_sections[idx]}")
```
